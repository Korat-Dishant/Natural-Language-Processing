###preprocessing function
it removes stop-words and punctuations and and return lammatized text

```python

import spacy

nlp = spacy.load("en_core_web_sm")

def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)

    return " ".join(filtered_tokens)
```

```python
text_corpus = df['headline'].apply(preprocess)
```

```python
text_corpus[0]
```

##Bag of words

You can use sklearn's countvectorizer which create features for text based on word accurance in data

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer.fit(text_corpus)
```

```python
vectorizer.vocabulary_
```

extrating features from our sarcasm data corpus using bag of words

```python
vector = vectorizer.transform(text_corpus)
print(vector.toarray())
```

##Tf-Idf

another way to extract features from text is tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
vectorized_corpus = vectorizer.fit_transform(text_corpus)
```
