# Tokenization


```python
import numpy as np
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import *
from nltk.stem.wordnet import WordNetLemmatizer
import spacy
from spacy import displacy
import re

```

```python
nltk.download('stopwords')
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
string = 'Sample String LDRP-itr , our website is www.ldrpitr.ac.in. we are glad that you are here. contact us at contact@ldrp.ac.in. Its situated in sector 15 gandhinagar '
print(string)
```

```python
doc = nlp(string)

for token in doc:
    print(token.text)
```

```python
len(doc.vocab)
```

```python
dir(doc)
```

```python
docu = nlp("Let's visit St. Louis in the U.S. next year.")
for item in docu:
    print(item)
```

# Named Entities Recognition


```python
for ent in doc.ents:
    print(ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
```

```python
for chunk in doc.noun_chunks:
    print(chunk.text)
```

## Visualizing dependency


```python
displacy.render(doc, style='dep', jupyter=True, options={'distance': 110})
```

## Visualizing NER

```python
displacy.render(doc, style='ent', jupyter=True)
```

#Regular Expressions


```python
pattern = '\w+'
result = re.findall(pattern, string)
print(result)
```

```python
for match in re.finditer("on",string):
    print(match.span())
```

```python
contactNumber = "Our cmpany's number is 5686536251"
phone = re.search(r'\d\d\d\d\d\d\d\d\d\d',contactNumber)
phone.group()
```

#Stemming

```python
p_stemmer = PorterStemmer()
```

```python
for token in doc:
    print(token.text+' --> '+p_stemmer.stem(token.text))
```

#Lemmatizatioin

```python
for token in doc:
    print(token.text, '\t', token.pos_, '\t', token.lemma, '\t', token.lemma_)
```

```python
def url_remove(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

text1 = string
print('without links:', url_remove(text1))
```

```python
stopwordslist = stopwords.words('english')
print(stopwordslist)
print('Total:',len(stopwordslist))
```

```python
stop_words = set(stopwords.words('english'))
result = []
for token in doc :
  if token not in stop_words :
    result.append(token)

print('cleaned string : ',result)

```
